## Facial keypoints detection



### Load and preprocess data

Script `get_data.py` unpacks data â€” images and labelled points. 6000 images are located in `images` folder and keypoint coordinates are in `gt.csv` file. Run the cell below to unpack data.

```python
### Useful routines for preparing data
from numpy import array, zeros
from os.path import join
from skimage.color import gray2rgb
from skimage.io import imread
from skimage.transform import resize

def load_imgs_and_keypoints(dirname='facial-keypoints'):
    # Write your code for loading images and points here
    df = pd.read_csv(os.path.join('data', 'gt.csv'))
    imgs = []
    df_picShape = pd.DataFrame(np.zeros((df.shape[0], 2)), columns=['x','y'])
    for i, pic in enumerate(df['filename']):
        img = imread(os.path.join('data', 'images', pic))
        #print(img.shape)
        df_picShape.loc[i] = img.shape[:2]
        img = resize(img, (100,100), mode='constant')
        imgs.append(gray2rgb(img))
        #print(img.shape)
        #print()
    df.iloc[:,2::2] = df.iloc[:,2::2].div(df_picShape['y'], axis=0) - 0.5
    df.iloc[:,1::2] = df.iloc[:,1::2].div(df_picShape['x'], axis=0) - 0.5
    imgs = np.stack(imgs)
    return imgs, df.iloc[:,1:].values

imgs, points = load_imgs_and_keypoints()
```

### Visualize data

```python
import matplotlib.pyplot as plt
# Circle may be useful for drawing points on face
# See matplotlib documentation for more info
from matplotlib.patches import Circle

def visualize_points(img, points):
    # Write here function which obtains image and normalized
    # coordinates and visualizes points on image
    fig,ax = plt.subplots(1)
    ax.imshow(img)
    for i, j in zip(points[0::2],points[1::2]):       
        ax.add_patch(Circle(((i+0.5)*100, (j+0.5)*100), 2, color='r'))
    
visualize_points(imgs[1], points[1][:24])
```

```python
def flip_img(img, points):
    # Write your code for flipping here
    copy = points.copy()
    copy[0::2] = -(copy[0::2])
    # head swap
    for i in range(4)[::2]:
        copy[i], copy[8-i-2] = copy[8-i-2], copy[i] # x
        copy[i+1], copy[8-i-1] = copy[8-i-1], copy[i+1] # y
        
    # eye swap
    for i in range(6)[::2]:
        copy[8+i], copy[20-i-2] = copy[20-i-2], copy[8+i] # x
        copy[8+i+1], copy[20-i-1] = copy[20-i-1], copy[8+i+1] # y
        
    # mouth swap
    copy[22], copy[26] = copy[26], copy[22] # x
    copy[27], copy[23] = copy[23], copy[27] # y
        
    return np.fliplr(img), copy


f_img, f_points = flip_img(imgs[1], points[1])
visualize_points(f_img, f_points)
```

```python
# Write your code here
aug_imgs_train = np.zeros(imgs_train.shape)
aug_points_train = np.zeros(points_train.shape)
for i in range(imgs_train.shape[0]):
    f_img, f_points = flip_img(imgs_train[i], points_train[i])
    aug_imgs_train[i] = f_img
    aug_points_train[i] = f_points
```

```python
aug_imgs_train = np.vstack([aug_imgs_train, imgs_train])
aug_points_train = np.vstack([aug_points_train, points_train])
```

```python
from keras.models import Sequential
from keras.layers import (
    Conv2D, MaxPooling2D, Flatten,
    Dense, Dropout
)

model = Sequential()
# Define here your model

model.add(Conv2D(64, kernel_size=(3, 3), activation='relu', input_shape=(100, 100, 3)))
model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Conv2D(64, kernel_size=(3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
#model.add(Dropout(0.25))

model.add(Conv2D(128, (3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Conv2D(128, (3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
#model.add(Dropout(0.25))

model.add(Flatten())
model.add(Dense(512, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(28))
```
